### Initializing

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://egov.uscis.gov/processing-times/historic-pt"
html = requests.get(url).text
```

```python
soup = BeautifulSoup(html)
```

```python
table = soup.select("table.records")[0]
```

### Table header

```python
header_els = table.select("tr")[0]
```

```python
headers = header_els.select("th")
```

```python
header_list = []
for head in headers:
    head = head.text.strip()
    header_list.append(head)
```

```python
header_list
```

```python
header_list[8] = 'FY 2024'
```

```python
header_list
```

```python
theadels = soup.select('table thead')[0].select('th')
```

```python
head = []
for el in theadels:
    element = el.text.strip()
    head.append(element)
```

```python
head[8] = 'FY 2024'
```
