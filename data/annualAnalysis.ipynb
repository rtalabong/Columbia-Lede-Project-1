### Annual and per president

```python
import pandas as pd
```

```python
df = pd.read_csv('annualCleaned.csv')
```

```python
df.head()
```

```python
value_columns = df.columns[1:]
```

```python
df['Lowest'] = df[value_columns].min(axis=1)
df['Highest'] = df[value_columns].max(axis=1)
df['Average'] = df[value_columns].mean(axis=1).round(2)
df['Median'] = df[value_columns].median(axis=1)
```

```python
obama_columns = ['2013', '2014', '2015', '2016']
trump_columns = ['2017', '2018', '2019', '2020']
biden_columns = ['2021', '2022', '2023', '2024']
```

```python
df['Obama mean'] = df[obama_columns].mean(axis=1).round(2)
df['Trump mean'] = df[trump_columns].mean(axis=1).round(2)
df['Biden mean'] = df[biden_columns].mean(axis=1).round(2)
```

```python
df['Obama median'] = df[obama_columns].median(axis=1).round(2)
df['Trump median'] = df[trump_columns].median(axis=1).round(2)
df['Biden median'] = df[biden_columns].median(axis=1).round(2)
```

```python
df
```

```python
df.to_csv('annualAnalyzed_final.csv', index = False)
```

```python
columns_for_viz = ['Form','Obama mean', 'Trump mean', 'Biden mean']
df_for_viz = df[columns_for_viz]
```

```python
median_columns = ['Form', 'Obama median', 'Trump median', 'Biden median']
df_median = df[median_columns]
```

```python
df_for_viz.to_csv('annualAnalyzed_forviz.csv', index = False)
```

```python
df_median.to_csv('presidentsMedianFinal.csv', index = False)
```

### Biden quarterly

```python
na_values = ['N/A', 'NaN', ' N/A ', 'N/a']
df = df = pd.read_csv('bidenPerQuarterCleaned.csv', na_values=na_values)
```

```python
value_columns = df.columns[3:]
```

```python
columns_to_convert = ['2023Q4', '2023Q3', '2023Q2', '2023Q1', '2022Q4', '2022Q3', '2022Q1']
df[columns_to_convert] = df[columns_to_convert].astype(float)
```

```python
df.dtypes
```

```python
df['Lowest'] = df[value_columns].min(axis=1)
df['Highest'] = df[value_columns].max(axis=1)
df['Average'] = df[value_columns].mean(axis=1).round(2)
df['Median'] = df[value_columns].median(axis=1)
df['Highest quarter'] = df[value_columns].idxmax(axis=1)
df['Lowest quarter'] = df[value_columns].idxmin(axis=1)
```

```python
df
```

```python
df['Highest quarter'].value_counts()
```

```python
df['Lowest quarter'].value_counts()
```

```python
df.to_csv('bidenQuarterlyAnalyzed_final.csv', index = False)
```

```python
columns_for_viz = ['Form', 'Lowest', 'Highest', 'Average', 'Median']
df_for_viz = df[columns_for_viz]
```

```python
df_for_viz.to_csv('bidenQuarterlyAnalyzed_forviz.csv', index = False)
```

```python
min_max_columns = ['Form', 'Lowest', 'Median', 'Highest']
df_minmax = df[min_max_columns]
df_minmax.to_csv('bidenQuarterlyAnalyzed_minmax.csv', index = False)
```
